```python
from scipy.stats import binom
import matplotlib.pyplot as plt
```

x = 2, n = 5, p = 0.5

```python
# PMF - Probability mass function or binomial formula
a = binom.pmf(2,5,0.5)
print(a.round(4))
```

```python
# CDF - Cummulative distributive function or P(x<=x)
b = binom.cdf(2,5,0.5)
print(b.round(4))
```

```python
mean = binom.mean(5,0.5)
var = binom.var(2,0.5)
print("Mean of Distribution = ",mean.round(4))
print("Variance of Distribution = ",var.round(4))
```

```python
n = 5
p = 0.5
x_values = list(range(n+1))
dist = [round(binom.pmf(x,n,p),4) for x in x_values]
print(dist)
print("Mean = ",round(binom.mean(n,p),4))
print("Variance = ",round(binom.var(n,p),4))

plt.stem(x_values,dist,markerfmt = "o",linefmt = "grey")
for i in range(len(x_values)):
    plt.annotate(str(dist[i]), xy=(x_values[i], dist[i]))

plt.xlabel("No of successes")
plt.ylabel("Binomial Distribution")
plt.show()
```
