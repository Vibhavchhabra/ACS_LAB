```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data = pd.read_excel("Items.xlsx")
```

```python
data.head()
```

```python
data["Item Name"].value_counts()
```

```python
plt.plot(data["Item Name"], data["Units Sold"])
```

```python
plt.pie(data["Units Sold"], labels = data["Item Name"])
plt.legend(data["Item Name"])
```

```python
min([13,24,56,77,89,90])
```

```python
plt.bar(data["Item Name"], data["Store 1"], label = "Store 1")
plt.bar(data["Item Name"], data["Store 2"],bottom =  data["Store 1"],label = "Store 1")

plt.xlabel("Item Name")
plt.ylabel("Quantity")
plt.legend()
plt.show()
```

```python
plt.barh(data["Item Name"], data["Store 1"], label = "Store 1")
plt.barh(data["Item Name"], data["Store 2"],left =  data["Store 1"],label = "Store 1")

plt.ylabel("Item Name")
plt.xlabel("Quantity")
plt.legend()
plt.show()
```

```python
width = 0.3
bar1 = np.arange(len(data["Item Name"]))
bar2 = [i + width for i in bar1]

plt.bar(bar1, data["Store 1"], width = width,label = "Store 1")
plt.bar(bar2, data["Store 2"],width = width,label = "Store 2")

plt.xlabel("Item Name")
plt.ylabel("Quantity")
plt.xticks(bar1 + width/2,data["Item Name"])
plt.legend()
plt.grid(color = "black", linewidth = 0.3, linestyle = "--")
plt.show()
```

```python
width = 0.3
bar1 = np.arange(len(data["Item Name"]))
bar2 = [i + width for i in bar1]

plt.barh(bar1, data["Store 1"],width,label = "Store 1")
plt.barh(bar2, data["Store 2"],width,label = "Store 2")

plt.ylabel("Item Name")
plt.xlabel("Quantity")
plt.yticks(bar1 + width/2,data["Item Name"])
plt.legend()
plt.grid(color = "black", linewidth = 0.3, linestyle = "--")
plt.show()
```
